```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Supongamos que tienes un DataFrame llamado df
df = pd.DataFrame({
    'DEPARTAMENTO': ['A', 'B', 'A', 'B', 'C'],
    'MUNICIPIO': ['X', 'Y', 'X', 'Y', 'Z'],
    'CIUDAD_DANE': [101, 102, 101, 103, 104],
    'TIPO_VEHICULO': ['Auto', 'Camión', 'Auto', 'Moto', 'Camión'],
    'FECHA_RECUPERACION': pd.to_datetime(['2022-01-01', '2022-01-02', '2022-01-03', '2022-01-04', '2022-01-05']),
    'CANTIDAD': [10, 15, 20, 25, 30]
})

# Visualización de variables categóricas
plt.figure(figsize=(15, 8))

plt.subplot(2, 2, 1)
sns.countplot(x='DEPARTAMENTO', data=df)

plt.subplot(2, 2, 2)
sns.countplot(x='TIPO_VEHICULO', data=df)

plt.subplot(2, 2, 3)
sns.countplot(x='MUNICIPIO', data=df)

plt.show()
```

```python
# Visualización de variables numéricas
plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(x='FECHA_RECUPERACION', y='CANTIDAD', data=df)

plt.subplot(1, 2, 2)
sns.boxplot(x='DEPARTAMENTO', y='CANTIDAD', data=df)

plt.show()
```

```python
# Estadísticas descriptivas
estadisticas_descriptivas = df[['CANTIDAD']].describe()

print("Estadísticas Descriptivas para CANTIDAD:")
print(estadisticas_descriptivas)
```

```python
# Tabla de contingencia para la relación entre DEPARTAMENTO y TIPO_VEHICULO
tabla_contingencia = pd.crosstab(df['DEPARTAMENTO'], df['TIPO_VEHICULO'])
print("Tabla de Contingencia:")
print(tabla_contingencia)

# Gráfico de barras apiladas para la relación entre DEPARTAMENTO y TIPO_VEHICULO
plt.figure(figsize=(10, 6))
tabla_contingencia.plot(kind='bar', stacked=True)
plt.title('Relación entre DEPARTAMENTO y TIPO_VEHICULO')
plt.xlabel('DEPARTAMENTO')
plt.ylabel('Count')
plt.show()
```

```python
# Tabla de contingencia para la relación entre DEPARTAMENTO y TIPO_VEHICULO
tabla_contingencia = pd.crosstab(df['DEPARTAMENTO'], df['TIPO_VEHICULO'])
print("Tabla de Contingencia:")
print(tabla_contingencia)

# Gráfico de barras apiladas para la relación entre DEPARTAMENTO y TIPO_VEHICULO
plt.figure(figsize=(10, 6))
tabla_contingencia.plot(kind='bar', stacked=True)
plt.title('Relación entre DEPARTAMENTO y TIPO_VEHICULO')
plt.xlabel('DEPARTAMENTO')
plt.ylabel('Count')
plt.show()
```

```python
# Convertir FECHA_RECUPERACION a formato de fecha
df['FECHA_RECUPERACION'] = pd.to_datetime(df['FECHA_RECUPERACION'])

# Crear un gráfico de línea para visualizar la tendencia temporal de CANTIDAD
plt.figure(figsize=(12, 6))
sns.lineplot(x='FECHA_RECUPERACION', y='CANTIDAD', data=df)
plt.title('Tendencia Temporal de CANTIDAD')
plt.show()
```

```python
# Boxplot para comparar CANTIDAD en diferentes DEPARTAMENTO
plt.figure(figsize=(10, 6))
sns.boxplot(x='DEPARTAMENTO', y='CANTIDAD', data=df)
plt.title('Comparación de CANTIDAD por DEPARTAMENTO')
plt.show()
```

```python
# Boxplot para comparar CANTIDAD en diferentes TIPO_VEHICULO
plt.figure(figsize=(10, 6))
sns.boxplot(x='TIPO_VEHICULO', y='CANTIDAD', data=df)
plt.title('Comparación de CANTIDAD por TIPO_VEHICULO')
plt.show()
```

```python
# Tabla de contingencia para DEPARTAMENTO y TIPO_VEHICULO
tabla_contingencia_categoricas = pd.crosstab(df['DEPARTAMENTO'], df['TIPO_VEHICULO'])
print("Tabla de Contingencia para DEPARTAMENTO y TIPO_VEHICULO:")
print(tabla_contingencia_categoricas)
```

```python
# Estadísticas descriptivas para CANTIDAD
estadisticas_cantidad = df['CANTIDAD'].describe()
print("Estadísticas Descriptivas para CANTIDAD:")
print(estadisticas_cantidad)

# Estadísticas descriptivas para FECHA_RECUPERACION (como números)
estadisticas_fecha = df['FECHA_RECUPERACION'].astype(int).describe()
print("\nEstadísticas Descriptivas para FECHA_RECUPERACION:")
print(estadisticas_fecha)
```

```python
# Correlación entre FECHA_RECUPERACION y CANTIDAD
correlacion_fecha_cantidad = df[['FECHA_RECUPERACION', 'CANTIDAD']].corr().iloc[0, 1]
print(f"\nCorrelación entre FECHA_RECUPERACION y CANTIDAD: {correlacion_fecha_cantidad:.2f}")
```
